```python
def load_graph_from_csv(file_path):
    df = pd.read_csv(file_path)
    G = nx.from_pandas_edgelist(df, source='source', target='target', edge_attr=True)
    print(f"Loaded graph from {file_path}")
    print("Number of nodes:", G.number_of_nodes())
    print("Number of edges:", G.number_of_edges())
    print("\n")
    return G
```

```python
all_books_path = 'data/hp_all_edges.csv'
book1_path = 'data/hp_book1_edges.csv'
book2_path = 'data/hp_book2_edges.csv'
book3_path = 'data/hp_book3_edges.csv'
book4_path = 'data/hp_book4_edges.csv'
book5_path = 'data/hp_book5_edges.csv'
book6_path = 'data/hp_book6_edges.csv'
book7_path = 'data/hp_book7_edges.csv'
```

```python
G_book1 = load_graph_from_csv(book1_path)
G_book2 = load_graph_from_csv(book2_path)
G_book3 = load_graph_from_csv(book3_path)
G_book4 = load_graph_from_csv(book4_path)
G_book5 = load_graph_from_csv(book5_path)
G_book6 = load_graph_from_csv(book6_path)
G_book7 = load_graph_from_csv(book7_path)
G_all_books = load_graph_from_csv(all_books_path)
```

```python
def graph_summaries(graph):
    num_nodes = graph.number_of_nodes()
    num_edges = graph.number_of_edges()
    average_degree = sum(dict(graph.degree()).values())/num_nodes
    density = nx.density(graph)
    num_connected_component = nx.number_connected_components(graph)
    average_clustering_coeffcient = nx.average_clustering(graph)

    return{
        "Number of Nodes": num_nodes,
        "Number of Edges": num_edges,
        "Average Degree": average_degree,
        "Density": density, 
        "Number of Connected Components": num_connected_component,
        "Average Clustering Coefficient": average_clustering_coeffcient
    }
```

```python
print(graph_summaries(G_all_books))
print(graph_summaries(G_book1))
print(graph_summaries(G_book2))
print(graph_summaries(G_book3))
print(graph_summaries(G_book4))
print(graph_summaries(G_book5))
print(graph_summaries(G_book6))
print(graph_summaries(G_book7))
```

```python
def centrality_metrics(graph):
    deg_cent = nx.degree_centrality(graph)
    weighted_deg_cent = dict(graph.degree(weight = 'weight'))
    bet_cent = nx.betweenness_centrality(graph)
    close_cent = nx.closeness_centrality(graph)
    eigenvec_cent = nx.eigenvector_centrality(graph)
    page_rank = nx.pagerank(graph)

    return{
        "Degree Centrality": deg_cent,
        "Weighted Degree Centrality": weighted_deg_cent, 
        "Betweenness Centrality": bet_cent,
        "Closeness Centrality": close_cent, 
        "Eigenvector Centrality": eigenvec_cent, 
        "Page Rank": page_rank
        
    }
```

```python
print(centrality_metrics(G_all_books))
```

```python
print(centrality_metrics(G_book1))
print(centrality_metrics(G_book2))
print(centrality_metrics(G_book3))
print(centrality_metrics(G_book4))
print(centrality_metrics(G_book5))
print(centrality_metrics(G_book6))
print(centrality_metrics(G_book7))
```

```python
def centrality_metrics_to_csv(graph, filename):
    # Calculate centrality metrics
    centrality_metrics = {
        "Degree Centrality": nx.degree_centrality(graph),
        "Weighted Degree Centrality": dict(graph.degree(weight='weight')),
        "Betweenness Centrality": nx.betweenness_centrality(graph),
        "Closeness Centrality": nx.closeness_centrality(graph),
        "Eigenvector Centrality": nx.eigenvector_centrality(graph, max_iter=1000),
        "PageRank": nx.pagerank(graph)
    }
    
    # Convert the centrality metrics to a DataFrame
    df = pd.DataFrame(centrality_metrics)
    
    # Save the DataFrame to a CSV file
    df.to_csv(f'{filename}.csv')
    
    print(f"Centrality metrics saved to {filename}")

books_graphs = [G_all_books, G_book1, G_book2, G_book3, G_book4, G_book5, G_book6, G_book7]

# Names or numbers for each book, used to create filenames
book_names = ['All Books','Book1', 'Book2', 'Book3', 'Book4', 'Book5', 'Book6', 'Book7']

# Iterate over each graph and save its centrality metrics to a CSV file
for book_graph, book_name in zip(books_graphs, book_names):
    filename = f'{book_name}.csv'
    centrality_metrics_to_csv(book_graph, filename)
```

```python
def get_top_10_characters(centrality_dicts):
    top_10 = {}
    for measure, centrality_dict in centrality_dicts.items():
        sorted_characters = sorted(centrality_dict.items(), key=lambda x: x[1], reverse=True)[:10]
        top_10[measure] = sorted_characters
    return top_10

def aggregate_top_characters(all_books_top_10):
    aggregated_characters = set()
    for book in all_books_top_10:
        for measure in all_books_top_10[book]:
            characters = [character for character, _ in all_books_top_10[book][measure]]
            aggregated_characters.update(characters)
    return aggregated_characters

def plot_evolution(evolution_data):
    plt.figure(figsize=(12, 8))
    sns.lineplot(data=evolution_data, dashes=False)
    plt.title('Evolution of Character Centrality Across Books')
    plt.xlabel('Book')
    plt.ylabel('Centrality Measure')
    plt.xticks(ticks=range(len(evolution_data.index)), labels=evolution_data.index, rotation=45)
    plt.legend(title='Character', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig('testplot.png')
    plt.show()
```

```python
all_books_top_10 = {
    "All Books": get_top_10_characters(centrality_metrics(G_all_books)),
    "Book 1": get_top_10_characters(centrality_metrics(G_book2)),
    "Book 2": get_top_10_characters(centrality_metrics(G_book2)),
    "Book 3": get_top_10_characters(centrality_metrics(G_book3)),
    "Book 4": get_top_10_characters(centrality_metrics(G_book4)),
    "Book 5": get_top_10_characters(centrality_metrics(G_book5)),
    "Book 6": get_top_10_characters(centrality_metrics(G_book6)),
    "Book 7": get_top_10_characters(centrality_metrics(G_book7)),   
}
aggregated_characters = aggregate_top_characters(all_books_top_10)
```

```python
book_titles = [f"Book {i}" for i in range(1, 8)]
characters = list(aggregated_characters) 
evolution_df = pd.DataFrame(index=book_titles, columns=characters).fillna(0)
```

```python
books = [G_all_books, G_book1, G_book2, G_book3, G_book4, G_book5, G_book6, G_book7]
for book_num, graph in enumerate(books, start=1):
    centrality = centrality_metrics(graph)["Degree Centrality"]  # Assuming using Degree Centrality
    book_title = f"Book {book_num}"
    for character in characters:
        if character in centrality:
            evolution_df.at[book_title, character] = centrality[character]
```

```python
plot_evolution(evolution_df)
```

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import networkx as nx
import numpy as np

# Using adjacency matrix to perform spectral clustering
adjacency_matrix = nx.adjacency_matrix(G_all_books)
# Convert adjacency matrix to dense format
dense_adjacency_matrix = adjacency_matrix.todense()

sc = SpectralClustering(5, affinity='precomputed', n_init=100)
clusters = sc.fit_predict(dense_adjacency_matrix)

# Create a dictionary that maps nodes to their cluster
partition = {node: clusters[i] for i, node in enumerate(G_all_books.nodes())}

# Preparing the node size and label size based on centrality measures
node_size = [1000 * nx.pagerank(G_all_books).get(node, 0) for node in G_all_books.nodes()]
label_size = {node: 10 * nx.betweenness_centrality(G_all_books).get(node, 0) for node in G_all_books.nodes()}

# Visualize the clusters
pos = nx.spring_layout(G_all_books)
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G_all_books, pos, partition.keys(), node_size=node_size, cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G_all_books, pos, alpha=0.5)
nx.draw_networkx_labels(G_all_books, pos, font_size=8, labels=label_size)
plt.show()
```
